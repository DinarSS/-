```python
num = int(input("Введите натуральное число: "))
num_str = str(num)
last_digit = num_str[-1]  # Последняя цифра

count_3 = count_last = count_even = 0

for digit in num_str:
    count_3 += digit == '3'
    count_last += digit == last_digit
    count_even += int(digit) % 2 == 0

print("Количество цифр 3:", count_3)
print("Количество вхождений последней цифры:", count_last)
print("Количество четных цифр:", count_even)
```

```python
N = int(input("Введите натуральное число N: "))
S = 0

for n in range(N):
    S += (-1) ** n / (2 * n + 1)

print("Значение S:", S)
```

```python
epsilon = 0.001
S = 0
n = 1

while True:
    term = (-1) ** (n - 1) / (n * (2 * n + 1))
    if abs(term) < epsilon:
        break
    S += term
    n += 1

print("Сумма ряда с заданной точностью:", S)
```

```python
N = int(input("Введите количество чисел (N > 1): "))
numbers = list(map(float, input("Введите N чисел через пробел: ").split()))

if len(numbers) != N:
    print("Ошибка: количество введенных чисел не совпадает с N")
else:
    is_increasing = all(numbers[i] < numbers[i + 1] for i in range(N - 1))
    print(is_increasing)
```
