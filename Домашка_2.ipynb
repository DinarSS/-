```python
import math
def f(x):
    if x <= -3:
        return -x**2 - 1.1*x + 9
    else:
        if x == -9**(1/3):
            return "Не определено"
        return math.log(x + 3) / (x**3 + 9)
x = float(input("Введите x: "))
result = f(x)
print(f"f({x}) = {result}")
```

```python
import math
def solve_quadratic(a, b, c):
    if a == 0:
        raise ValueError("a не должно быть равно 0")
    D = b**2 - 4*a*c
    t = D >= 0
    if t:
        x1 = (-b + math.sqrt(D)) / (2 * a)
        x2 = (-b - math.sqrt(D)) / (2 * a)
    else:
        x1 = x2 = None
    return t, x1, x2
a, b, c = map(float, input("Введите a, b, c через пробел: ").split())
t, x1, x2 = solve_quadratic(a, b, c)
print(f"t = {t}, x1 = {x1}, x2 = {x2}")
```

```python
def is_palindrome(number):
    if 100 <= number <= 999:
        str_num = str(number)
        return str_num == str_num[::-1]
    else:
        raise ValueError("Число должно быть трехзначным")
num = int(input("Введите трехзначное число: "))
if is_palindrome(num):
    print("Число является палиндромом")
else:
    print("Число не является палиндромом")
```

```python
def is_under_attack(white, black, move, black_piece):
    xw, yw = white
    xb, yb = black
    xm, ym = move
    if not all(1 <= coord <= 8 for coord in (xw, yw, xb, yb, xm, ym)):
        raise ValueError("Координаты должны быть в диапазоне от 1 до 8")
    def king_moves(x, y):
        return {(x + dx, y + dy) for dx in [-1, 0, 1] for dy in [-1, 0, 1] if (dx, dy) != (0, 0)}
    def knight_moves(x, y):
        return {(x + dx, y + dy) for dx, dy in [(2, 1), (2, -1), (-2, 1), (-2, -1),
                                                (1, 2), (1, -2), (-1, 2), (-1, -2)]}
    def rook_moves(x, y):
        return {(x, i) for i in range(1, 9)} | {(i, y) for i in range(1, 9)}
    if black_piece == "knight":
        under_attack = move in knight_moves(xb, yb)
    elif black_piece == "rook":
        under_attack = move in rook_moves(xb, yb)
    else:
        raise ValueError("Некорректная черная фигура")
    return not under_attack and move in king_moves(xw, yw)
a, b, c, d, e, f = map(int, input("Введите координаты a b c d e f через пробел: ").split())
black_piece = input("Введите черную фигуру (knight/rook): ").strip().lower()

if is_under_attack((a, b), (c, d), (e, f), black_piece):
    print("Белая фигура может пойти на указанное поле, не попадая под удар.")
else:
    print("Ход невозможен или он попадает под удар черной фигуры.")
```
