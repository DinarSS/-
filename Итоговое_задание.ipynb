begin 10

```python
a = float(input("Введите первое ненулевое число (a): "))
b = float(input("Введите второе ненулевое число (b): "))
if a == 0 or b == 0:
    print("Ошибка: числа должны быть ненулевыми.")
else:
    a2 = a ** 2
    b2 = b ** 2
    sum_squares = a2 + b2
    diff_squares = a2 - b2
    prod_squares = a2 * b2
    div_squares = a2 / b2
    print("Сумма квадратов:", sum_squares)
    print("Разность квадратов:", diff_squares)
    print("Произведение квадратов:", prod_squares)
    print("Частное квадратов:", div_squares)
```

for 23

```python
X = float(input("Введите вещественное число X: "))
N = int(input("Введите целое число N (> 0): "))
sum_series = X
for n in range(1, N + 1):
    power = X ** (2 * n + 1)
    factorial = 1
    for i in range(1, 2 * n + 2):
        factorial *= i
    sum_series += (-1) ** n * power / factorial
print(f"Приближенное значение sin(X) для X = {X} и N = {N}: {sum_series:.6f}")
```
