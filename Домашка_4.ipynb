```python
from collections import Counter

def prime_factorization(n):
    factors = []
    i = 2
    while i * i <= n:
        while n % i == 0:
            factors.append(i)
            n //= i
        i += 1
    if n > 1:
        factors.append(n)
    factor_counts = Counter(factors)
    return " * ".join(f"{factor}^{factor_counts[factor]}" if factor_counts[factor] > 1 else str(factor) for factor in sorted(factor_counts))
num = 200
print(f"{num} = {prime_factorization(num)}")
```

```python
def max_adjacent_elements_v2(arr):
    B = []
    for i in range(1, len(arr)):
        B.append(arr[i-1] if arr[i-1] > arr[i] else arr[i])
    return B
A = [7, 2, 9, 3, 6, 1, 8]
B = max_adjacent_elements_v2(A)
print(f"B = {B}")
```

```python
import math
def factorial_sum(n):
    total_sum = 0
    sign = 1
    for i in range(2, n+1, 2):
        total_sum += sign * math.factorial(i)
        sign *= -1
    return total_sum
n = 20
result = factorial_sum(n)
print(f"Сумма для n={n}: {result}")
```
