```python
def SumRange(A, B):
    if A > B:
        return 0
    return (B - A + 1) * (A + B) // 2
A = int(input("Введите A: "))
B = int(input("Введите B: "))
C = int(input("Введите C: "))
sum_AB = SumRange(A, B)
sum_BC = SumRange(B, C)
print(f"Сумма от A до B: {sum_AB}")
print(f"Сумма от B до C: {sum_BC}")
```

```python
def prime_factors(n):
    factors = []
    count = 0
    while n % 2 == 0:
        n //= 2
        count += 1
    if count > 0:
        factors.append((2, count))
    divisor = 3
    while divisor * divisor <= n:
        count = 0
        while n % divisor == 0:
            n //= divisor
            count += 1
        if count > 0:
            factors.append((divisor, count))
        divisor += 2
    if n > 1:
        factors.append((n, 1))
    return factors
def print_prime_factors(n):
    factors = prime_factors(n)
    result = []
    for factor, exponent in factors:
        if exponent > 1:
            result.append(f"{factor}^{exponent}")
        else:
            result.append(str(factor))
    print(" * ".join(result))
N = int(input("Введите натуральное число: "))
print_prime_factors(N)
```

```python
def create_b_array(A):
    B = []
    for i in range(len(A) - 1):
        B.append(max(A[i], A[i + 1]))
    return B
A = [1, 3, 5, -2, 0, 4, 0]
B = create_b_array(A)
print("Массив A:", A)
print("Массив B:", B)
```

```python
import math
def factorial(n):
    return math.factorial(n)
def compute_sum(n):
    total = 0
    for i in range(2, n + 1, 2):
        if (i // 2) % 2 == 0:
            total += factorial(i)
        else:
            total -= factorial(i)
    return total
n = int(input("Введите значение n (четное, не больше 20): "))
if n <= 20 and n % 2 == 0:
    result = compute_sum(n)
    print("Результат вычисления суммы:", result)
else:
    print("Некорректный ввод. Пожалуйста, введите четное число, не превышающее 20.")
```
